```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# grad dataset from this url
data = pd.read_csv("https://gist.githubusercontent.com/nstokoe/7d4717e96c21b8ad04ec91f361b000cb/raw/bf95a2e30fceb9f2ae990eac8379fc7d844a0196/weight-height.csv")

# Display the first few rows of the DataFrame
data.head()

```

```python
data.describe()
```

```python
import seaborn as sns
sns.scatterplot(data, x='Height', y="Weight", alpha=0.8)
```

```python
x = data[["Height"]]
x
```

```python
y = data[["Weight"]]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
model = LinearRegression()
model
```

```python
model.fit(x_train, y_train)
model
```

```python
import numpy as np

x_test.head(5)
```

```python
model.predict(x_test.head(5))
```

```python
y_test.head(5)
```

```python
y_test_prediction = model.predict(x_test)
y_test_prediction
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
rmse = root_mean_squared_error(y_test, y_test_prediction)
print(f"RMSE: {rmse}")
```
